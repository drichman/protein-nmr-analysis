### Common functions

```python
ab_up = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
ab_lo = ab_up.lower()

def sparky_lt(f):
    """Read Sparky "lt" list, return parameters indexed by residues.
    
    Parameters
    ----------
    f : string filename
    
    Returns
    -------
    Series of heights as integers, indexed by amino acid labels
    """
    data = pd.read_table(f, delim_whitespace=True, squeeze=True,
        index_col='Assignment').drop('Height',axis=1).dropna()
    if '?-?' in data.index:
        data.drop('?-?', inplace=True)
    resnames = [i.rstrip('NH-') for i in data.index]
    data.index = resnames
    data.columns=['wN','wH','height']
    return data

from re import findall
extract_pH = lambda x: findall('[0-9]p[0-9]', x)[0].replace('p', '.')

def df_lt_build(loc, variant_id):
    """Construct DataFrame with peak list parameters for multiple conditions

    Parameters
    ----------
    loc : string directory name containing peak list files
    variant_id : string, part of the peak list file names
    
    Returns
    -------
    DataFrame with MultiIndex columns: peak list params at each pH 
    """
    lt_files = [loc+i for i in listdir(loc) if i.startswith(variant_id)]
    keys = [extract_pH(i) for i in lt_files]
    df_lt = pd.concat([sparky_lt(f) for f in lt_files], axis=1, keys=keys)
    print(variant_id+': '+str(keys)) # just so we can read the pH conditions
    return df_lt

def int_index(df):
    ints = [int(i.strip(ab_up+ab_lo+'-')) for i in df.index]
    df.index = ints
    return df.sort_index()

def pymol_list(res_list):
    """Produces '1+2+3...' list format to be copied
    to PyMOL for coloring or grouping by 'resi'.
    """
    a = ''
    for i in res_list:
        a+=str(i)
        a+='+'
    return a
```

```python
def goes_missing(from_here, to_here):
    """Make list of residues in 'from' peak list missing in 'to' peak list.
    
    Parameters
    ----------
    from_here : DataFrame
    to_here : DataFrame
    
    Returns
    -------
    list of residue numbers
    """
    missing = list(set(from_here.dropna().index).difference(set(to_here.dropna().index)))
    return missing
```

```python
def pair_iterate(df):
    """Applies goes_missing to successive pairs of pH values,
    
    ie if a protein was studied at pH 8, 7, 6, 5, 4, this function
    makes comparisons between condition pairs 8 and 7, 7 and 6,
    6 and 5, 5 and 4.
    
    Parameters
    ----------
    df : DataFrame of peak list parameters for each pH studied
    df_dphs: column of 
    
    Returns
    -------
    dictionary eg { from pH to pH : residues that go missing }
    """
    phlist = df.columns.levels[0][::-1]
    missing_dict = {}
    
    start_missing = goes_missing(df_lt_dphs['7.2'],df[phlist[0]])
    missing_dict[phlist[0]] = start_missing
    
    for i, p in enumerate(phlist[:-1]):
        new_missing = goes_missing(df[p],df[phlist[i+1]])
        cumulative = missing_dict[p] + new_missing
        missing_dict[phlist[i+1]] = cumulative
    for k, v in missing_dict.items():
        if v:
            missing_dict[k] = sorted(v)
        else:
            missing_dict[k] = []
    return missing_dict
```

Note that lists of resonances that have "gone missing" from one pH to the next generated by this function are not cumulative, but that final visualizations of missing residues at each pH will be made to reflect the cumulative effect of broadening from all previous pH conditions.

```python
brd_ph_dphs = pair_iterate(df_lt_dphs)
brd_ph_l25k = pair_iterate(df_lt_l25k)
brd_ph_l125k = pair_iterate(df_lt_l125k)
```

```python
def brd_lists(brd_dict):
    for ph,res in brd_dict.items():
        print(ph+'\n', file=out_brd_f)
        print(str(res)+'\n', file=out_brd_f)
        print(pymol_list(res)+'\n', file=out_brd_f)
```

L25K: relaxation dispersion CPMG frequency series from 600 MHz and 800 MHz spectrometers, at pH 6.3. After processing and resonance assignment of each spectrum, Sparky peak lists served as input for relaxation dispersion model fitting by the [relax software](http://www.nmr-relax.com). L125K: relaxation dispersion CPMG frequency "endpoints"--max and min CPMG frequencies--from 600 MHz and 800 MHz spectrometers, at pH 7.2 and 5.2. After processing and resonance assignment of the endpoint spectra, peak heights were used to estimate $R_{ex}$ for each backbone amide.

```python
def est_endpts(few_loops, many_loops, ct_time):
    """Calculate the intensity ratio of CPMG extremes and make Rex estimate.
    
    Parameters
    ----------
    few_loops : Series with peak height data vs residue
    many_loops : Series with peak height data vs residue
    ct_time : float representing constant-time relaxation period in ms
    
    Returns
    -------
    Series containing Rex estimates indexed by residues
    """
    ratio = few_loops/many_loops #low_peak_intensity/high_peak_intensity
    rex_est = -np.log(ratio)/(ct_time*1e-3)
    return rex_est
```

```python
rex_l25k_600 = est_endpts(l25k_600_few, l25k_600_many, 28.8)
rex_l25k_800 = est_endpts(l25k_800_few, l25k_800_many, 28.8)
```

```python
def parse_relax(f):
    """Get data from relax output file into DataFrame.
    """
    relax_cols = ['mol_name','res_num','res_name','spin_num','spin_name','value','error']
    unnec = ['mol_name', 'res_name', 'spin_num', 'spin_name']
    data = pd.read_table(f, delim_whitespace=True, 
        skiprows=3, header=None, names=relax_cols,
        index_col='res_num').drop(unnec, axis=1)
    data.replace(to_replace='None',value=np.nan,inplace=True)
    return data
```

```python
extract_field = lambda x: findall('[0-9][0][0]', x)[0]
extract_loops = lambda x: findall('.[0-9]\.', x)[0].rstrip('.').lstrip('_')

def df_rd_end_build(loc, variant_id):
    """Construct DataFrame with peak list parameters for multiple conditions

    Parameters
    ----------
    loc : string directory name containing peak list files
    vaddriant_id : string, part of the peak list file names
    
    Returns
    -------
    DataFrame with MultiIndex columns: peak heights at each condition
    """
    lt_files = [loc+i for i in listdir(loc) if i.startswith(variant_id)]
    pH = [extract_pH(i) for i in lt_files]
    fields = [extract_field(i) for i in lt_files]
    loops = [extract_loops(i) for i in lt_files]
    tuples = list(zip(pH, fields, loops))
    multcols = pd.MultiIndex.from_tuples(tuples, names=['pH', 'field', 'loops'])
    df_lt = pd.concat([sparky_lt(f)['height'] for f in lt_files], axis=1)
    df_lt.columns = multcols
    return df_lt
```

##### Alpha parameter estimates

Taking the significant $R_{ex}$ values:

```python
rex_600_sig = rex_600[ rex_600 > rex_600.mean()+rex_600.std() ].dropna(how='all')
rex_800_sig = rex_800[ rex_800 > rex_800.mean()+rex_800.std() ].dropna(how='all')
```

```python
def alpha(rexB1, rexB2, B1=14.1, B2=18.8):
    """
    Parameters
    ----------
    rexB1, rexB2 : series containing Rex values indexed by residues
    B1, B2 : floats, magnetic fields in Tesla, B1 < B2
    
    Returns
    -------
    Series of alpha values indexed by residues
    """
    B_ratio = (B2+B1)/(B2-B1)
    alpha = B_ratio*(rexB2 - rexB1)/(rexB2 + rexB1)
    return alpha.sort_index()
```

This is what the $\alpha$ parameter means:

$0 < \alpha < 1$ : slow exchange

$\alpha = 1$ : intermediate exchange

$1 < \alpha < 2$ : fast exchange

```python
alph = alpha(rex_600_sig, rex_800_sig)
plt.rcParams['font.size']=14
alph.plot(kind='bar', figsize=(18,5));
plt.xticks(rotation='horizontal');
plt.xlabel('Residue');
#plt.ylabel();
```

```python
slow_exchange = alph[ (alph > 0) & (alph < 0.8) ].dropna(how='all')
int_exchange = alph[ (alph > 0.8) & (alph < 1.2) ].dropna(how='all')
fast_exchange = alph[ (alph > 1.2) & (alph < 2) ].dropna(how='all')
```

```python
alpha_table = pd.concat([slow_exchange,int_exchange,fast_exchange], keys=['slow','int','fast'])
alpha_table.index.names = ['regime','res']
alpha_table.to_csv(dir_rd_ana+'l125k_alpha_table.csv', float_format='%.1f')
```

Available information:

* CR72 model: $k_{ex}$ (and hence $\tau_{ex}$), $P_A$, $\Delta\omega_N$; $k_{AB}$ can be derived since $k_{AB}=P_B k_{ex}$
* TSMFK01 model: $k_{AB}$, $\Delta\omega_N$; if $P_B$ can be measured from A- and B-state peak heights then $k_{ex}$ can be derived since $k_{ex}=k_{AB}/P_B$, 

```python
kex,tex,kab,pa,pb,dw,chi2 = \
    [parse_relax(f).dropna(how='all').astype(float) for f in rlx_par_files[1:]]
```

```python
def par_model_select(df, df_models, modelname):
    """Pull model-specific parameters from parameter DataFrame.
    
    Parameters
    ----------
    df : DataFrame with parameter values
    df_models : DataFrame with relax model selection results
    modelname: string 'CR72' or 'TSMFK01' to select model
    
    Returns
    -------
    DataFrame with parameter values only for a specific model
    """
    index_subset = df_models[df_models['value'].str.contains(modelname)].index
    return df.ix[index_subset]
```

## Characterization of conformational exchange by backbone 15N ZZ-exchange

Of all the residues showing cross-peaks in (H,H), (H,N), and (H,H,N) ZZ-exchange spectra, a subset with auto- and cross-peaks of good resolution could be used for quantitative measurements of decay and build-up curves.

```python
def sparky_rh(f):
    """Read Sparky rh file, return DataFrame.
    
    Parameters
    ----------
    f : string filename
    
    Returns
    -------
    DataFrame indexed by timepoints; columns are peak labels
    """
    data = pd.read_table(f, delim_whitespace=True,
        index_col='Assignment').drop(['T-decay','SD'], axis=1)
    timepoints = timepoint_parse(data.columns)
    data.columns = timepoints
    return data.T

def timepoint_parse(ind):
    """Parse time values from Sparky rh column format.
    
    Parameters
    ----------
    ind : list (ie from index or column of DataFrame)
    
    Returns
    -------
    List of time values as floats
    """
    t_list = []
    for i in ind:
        time_str = i.split('/')[0]
        try:
            t_list.append(float(time_str))
        except ValueError:
            t_list.append(float(time_str.split('_')[0]))
    return t_list
```

```python
df_rh_l25k_zz = sparky_rh(dir_zz_raw+'l25k_zz_t_series_rh.txt')
```

```python
regex_aa = '[A-Z]{3}[0-9]{2,3}N-HN'
regex_bb = '[a-z]{3}[0-9]{2,3}N-HN'
regex_ab = '[A-Z]{3}[0-9]{2,3}N-[a-z]{3}[0-9]{2,3}HN'
regex_ba = '[a-z]{3}[0-9]{2,3}N-[A-Z]{3}[0-9]{2,3}HN'

regex_list = [regex_aa, regex_bb, regex_ab, regex_ba]
df_l25k_zz_aa, df_l25k_zz_bb, df_l25k_zz_ab, df_l25k_zz_ba = [df_rh_l25k_zz.filter(regex=re) for re in regex_list]

for df in [df_l25k_zz_aa, df_l25k_zz_bb, df_l25k_zz_ab, df_l25k_zz_ba]:
    resnums = [col.split('-')[0].strip(ab_up+ab_lo) for col in df.columns]
    df.columns = resnums
```

```python
def a11(ra, kab):
    return ra + kab

def a22(rb, kba):
    return rb + kba

def lamb1(ra, rb, kab, kba):
    asum = a11(ra,kab)+a22(rb,kba)
    adif = a11(ra,kab)-a22(rb,kba)
    return .5 * ( asum + np.sqrt( adif**2 + 4*kab*kba) )

def lamb2(ra, rb, kab, kba):
    asum = a11(ra,kab)+a22(rb,kba)
    adif = a11(ra,kab)-a22(rb,kba)
    return .5 * ( asum - np.sqrt( adif**2 + 4*kab*kba) )

def exp(lamb, t):
    return np.exp(-lamb*t)

def iaa(t, iaa0, ra, rb, kab, kba):
    lamb2a11 = lamb2(ra,rb,kab,kba) - a11(ra,kab)
    lamb1a11 = lamb1(ra,rb,kab,kba) - a11(ra,kab)
    lambdiff = lamb1(ra,rb,kab,kba) - lamb2(ra,rb,kab,kba)
    return iaa0 * ( ( -lamb2a11*exp(lamb1(ra,rb,kab,kba),t) + lamb1a11*exp(lamb2(ra,rb,kab,kba),t) ) / lambdiff )

def ibb(t, ibb0, ra, rb, kab, kba):
    lamb2a22 = lamb2(ra,rb,kab,kba) - a22(ra,kab)
    lamb1a22 = lamb1(ra,rb,kab,kba) - a22(ra,kab)
    lambdiff = lamb1(ra,rb,kab,kba) - lamb2(ra,rb,kab,kba)
    return ibb0 * ( ( -lamb2a22*exp(lamb1(ra,rb,kab,kba),t) + lamb1a22*exp(lamb2(ra,rb,kab,kba),t) ) / lambdiff )

def iab(t, iaa0, ra, rb, kab, kba):
    lambdiff = lamb1(ra,rb,kab,kba) - lamb2(ra,rb,kab,kba)
    return iaa0 * ( ( -kab*exp(lamb1(ra,rb,kab,kba),t) + kab*exp(lamb2(ra,rb,kab,kba),t) ) / lambdiff )

def iba(t, ibb0, ra, rb, kab, kba):
    lambdiff = lamb1(ra,rb,kab,kba) - lamb2(ra,rb,kab,kba)
    return ibb0 * ( ( -kba*exp(lamb1(ra,rb,kab,kba),t) + kba*exp(lamb2(ra,rb,kab,kba),t) ) / lambdiff )

def zz_eq(t, iaa0, ibb0, ra, rb, kab, kba):
    i_aa = iaa(t, iaa0, ra, rb, kab, kba)
    i_bb = ibb(t, ibb0, ra, rb, kab, kba)
    i_ab = iab(t, iaa0, ra, rb, kab, kba)
    i_ba = iba(t, ibb0, ra, rb, kab, kba)
    return np.concatenate([i_aa, i_bb, i_ab, i_ba])
```

```python
from lmfit import Model, Parameter

zz_mod = Model(zz_eq)
guess = dict(iaa0=1., ibb0=.3, ra=Parameter(value=1.,min=0), rb=Parameter(value=1.,min=0), kab=5., kba=5.)
timepoints = df_rh_l25k_zz.index.values
N = len(timepoints)
```

```python
def zz_fit_plotting(reslist, zz_labels, zz_markers, zz_styles, ftxt, ffig, fcsv, mode='AB'):
    """
    reslist : list of residue numbers (strings)
    zz_labels : list of names for curves (should be length 4)
    zz_colors : list of matplotlib color (should be length 4)
    fname_txt : string filename
    fname_fig : string filename
    mode : 'AB' when peak heights of AA > BB
           'BA' when peaks heights of BB > AA
    """
    nrows = int(np.ceil(len(reslist) / 3.))
    zz_fit_fig, zz_subplots = plt.subplots(nrows, 3, figsize=(14,8))
    zz_subplots.flat[-1].set_visible(False)
    
    zz_k_results_dict = {}
    zz_fit_report = open(dir_zz_ana+ftxt, 'w')
    
    for ax, res in zip(zz_subplots.flat, reslist):
        aavals=df_l25k_zz_aa[res].values
        bbvals=df_l25k_zz_bb[res].values
        abvals=df_l25k_zz_ab[res].values
        bavals=df_l25k_zz_ba[res].values
        if mode == 'AB':
            aarsc = aavals/aavals[0]
            bbrsc = bbvals/aavals[0]
            abrsc = abvals/aavals[0]
            barsc = bavals/aavals[0]
        if mode == 'BA':
            aarsc = aavals/bbvals[0]
            bbrsc = bbvals/bbvals[0]
            abrsc = abvals/bbvals[0]
            barsc = bavals/bbvals[0]

        zz_data = np.concatenate([aarsc,bbrsc,abrsc,barsc])    
        result = zz_mod.fit(zz_data, t=timepoints, **guess)

        # drawing results to the res-specific subplot

        leg_points = []
        leg_curves = []
        for i,data in enumerate([aarsc,bbrsc,abrsc,barsc]):
            points, = ax.plot(timepoints, data, zz_markers[i]+'k')
            curve, = ax.plot(timepoints, result.best_fit[i*N:(i+1)*N], zz_styles[i]+'k');
            leg_points.append(points)
            leg_curves.append(curve)
        ax.set_title(res, fontsize=14)

        # saving the res-specific results in text formats
        
        zz_k_results_dict[res] = [ result.values['kab'], result.values['kba'], \
                             result.values['kab']+result.values['kba'] ]
        
        print(res+'\n', file=zz_fit_report)
        print(result.fit_report(show_correl=False)+'\n', file=zz_fit_report)
        
    # full plot settings and output
    
    zz_fit_fig.legend(leg_points, zz_labels, loc=(0.705,0.3))
    zz_fit_fig.legend(leg_curves, zz_labels, loc=(0.8,0.3))
    zz_fit_fig.text(0.5, 0.05, 'Mixing time (s)', ha='center', fontsize=14)
    zz_fit_fig.text(0.075, 0.5, 'Peak height (scaled)', va='center', rotation='vertical', fontsize=14)    
    zz_fit_fig.savefig(dir_zz_fig+ffig)
    
    # full text outputs
    
    df_zz_k_results = pd.DataFrame(zz_k_results_dict, index=['kab','kba','kex']).T
    df_zz_k_results.to_csv(dir_zz_ana+fcsv, float_format='%.1f')
    zz_fit_report.close()
```

```python
def alt_derived(df):
    """Derive chem shift diffs and populations from peak lists.
    
    Parameters
    ----------
    df : DataFrame representing the parsed peak list
    
    Returns
    -------
    df : new DataFrame with the derived quantities
    """
    bpeaks = df.ix[[a for a in df.index if a.isupper()==False]]
    # take A-state peaks corresponding to detectable B-state peaks:
    apeaks = df.ix[[i.upper() for i in bpeaks.index]]
    aachar = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
    apeaks.index = [int(i.lstrip(aachar)) for i in apeaks.index]
    bpeaks.index = [int(i.lstrip(aachar.lower())) for i in bpeaks.index]

    dwN = np.round(abs(bpeaks['wN'] - apeaks['wN']), 1)
    dwH = np.round(abs(bpeaks['wH'] - apeaks['wH']), 1)

    total = apeaks['height'] + bpeaks['height']
    pa = np.round(apeaks['height']/total, 2)
    pb = np.round(bpeaks['height']/total, 2)

    return pd.concat([dwN,dwH,pa,pb], axis=1, keys=['dw','dwH','pA','pB'])
```

## Toy model of ionization and conformation equilibria

This uses the mathematics of coupled ionization and conformation equilibria (a thermodynamic cycle) to get a sense of what the microstate pKa values would need to be, given some observations and assumptions:

Equations derived as in Di Russo et al. (Roitberg lab) PLoS Comp Bio 2012, but terms are named differently; here I use:
* KaA, KaB for the acid dissociation constants in state A, B respectively
* Kop for the equilibrium constant between deprotonated conformations ([B]/[A]) ('K for opening reaction')
* KopH for the equilibrium constant between protonated conformations ([BH]/[AH])

```python
RT = 1.9872e-3*298
```

```python
def Kop(DG): # ∆G between conformations when deprotonated
    return np.exp(-DG/RT)

def KopH(DGH): # ∆G between conformations when protonated
    return np.exp(-DGH/RT)

def KaA(pKaA):
    return 10**(-pKaA)

def KaB(pKaB):
    return 10**(-pKaB)

def concH(pH):
    return 10**(-pH)

def nested(Kop, KaA, KaB):
    return Kop/KaB + 1/KaA

def concA(concH, Kop, KaA, KaB):
    return 1 / (1 + Kop + concH*nested(Kop, KaA, KaB))

def concAH(concH, Kop, KaA, KaB):
    return concH / ( KaA * (1 + Kop + concH*nested(Kop, KaA, KaB)) )

def concB(concH, Kop, KaA, KaB):
    return Kop / (1 + Kop + concH*nested(Kop, KaA, KaB))

def concBH(concH, Kop, KaA, KaB):
    return concH*Kop / ( KaB * (1 + Kop + concH*nested(Kop, KaA, KaB)) )

def pKa_app(Kop, pKaA, pKaB): # apparent pKa
    return pKaB + pKaA - np.log10( (Kop+1) / ( Kop*KaA(pKaA)+KaB(pKaB) ) )
```

### L25K

Properties and assumptions:
* L25K ∆G @ pH 7.5 = 4.6 kcal/mol
* L25K ∆G @ pH 5.5 = 2.7 kcal/mol
* L25K apparent pKa = 6.3; assume B-state pKa = 10.4 or close to it

So the ∆G between A and B states or ensembles is probably less than 4 kcal/mol and maybe even less that 2 even at pH 7.5.

$K_{op} = [B]/[A]$ (conformational equilibrium under the deprotonated condition particularly relevant at high pH) should probably be < 1 (associated ∆G > 0).

$K_{opH} = [BH]/[AH]$ (conformational equilibrium under the protonated condition particularly relevant at low pH) should probably be > 1 (associated ∆G < 0).

This is the reverse of the above Di Russo scenario.
