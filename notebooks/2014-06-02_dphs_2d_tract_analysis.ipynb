# Q: Can we rule out transient oligomerization?
# A: No.

I approach this question by comparing TRACT-derived rotational correlation times at high and low protein concentrations and between pH conditions. If transient oligomerization occurs at pH 7.4, we expect to observe a higher correlation time at pH 7.4 in the 0.75 mM sample compared to the 0.16 mM sample, while the correlation time should not change significantly at pH 3.3 between the 0.75 mM and 0.16 mM samples.

## Systematic fitting of TROSY and anti-TROSY decay series

The point is to find the roots $\tau_c$ of the equation
$$ R_{anti-TROSY} - R_{TROSY} = 2\eta_{xy} = 2p\delta_N \big(4J(0) + 3J(\omega_N)\big) (3cos^2\theta-1) $$
where
$$ p = \frac{\mu_0 \gamma_H \gamma_N h}{16 \pi^2 \sqrt{2} r^3_{HN}} $$

$$ \delta_N = \frac{ \gamma_N B_0 \Delta\delta_N }{3\sqrt{2}} $$

$$ J(\omega) = \frac{0.4 \tau_c}{1 + (\tau_c \omega)^2} $$

This simplifies to a polynomial to the 3rd power in $\tau_c$.

```python
import numpy as np
from scipy import optimize

mu = 4*np.pi*1e-7 # T^2 m^3 / J
gH = 267.522e6 # 1 / T s
gN = -27.126e6 # 1 / T s
oH = 2*np.pi*600.53e6 # radians / s (Brew2)
oN = 0.101329118*oH # based on IUPAC N/H gyromag ratio
h = 6.62608e-34 # J s
r = 1.02e-10 # m
B = oH/gH # T
DdN = 160.0e-6
theta = 24*np.pi/180 # radians
p0 = [1e7, 12.0] # Initial guesses for curve fitting

p = mu*gH*gN*h / (16*np.pi**2 * 2**.5 * r**3)
dN = gN*B*DdN / (3 * 2**.5)
const = 2*p*dN*(3 * np.cos(theta)**2 - 1)
```

```python
def tract_poly(x, twoeta):
    """x : independent variable representing tau_c
       twoeta : constant fixed by data fitting
    """
    return 1.6*const*oN**2*x**3 - twoeta*oN**2*x**2 + 2.8*const*x - twoeta

def tract_R(df, mode, p0):
    """Fits one decay for one residue.
    
    Parameters
    ----------
    df : DataFrame, actually df[res] from larger DataFrame
    mode : string 't' or 'a' to fit TROSY or anti-TROSY data
    p0 : initial guess array of format [a0, k0]
    
    Returns
    -------
    Decay rate (units 1/s)
    """
    t = np.asarray(df.ix[mode].index.values, dtype=float)
    y = np.asarray(df.ix[mode].values, dtype=float)
    popt, pcov = optimize.curve_fit(nmrfn.decay, xdata=t, ydata=y, p0=p0)
    return popt[1]

def calc_tau(res):
    twoeta = tract_R(res,'a',p0) - tract_R(res,'t',p0)
    sol = optimize.root(tract_poly, [10.], args=(twoeta))
    for x in sol.x:
        if np.isreal(x):
            tau_c = np.real(x)*1e9
        else:
            pass
    return tau_c
        
def single_sample(df):
    trimmed = df.dropna(axis=1)
    return trimmed.apply(calc_tau)

def sample_set(samp_list):
    mult = [single_sample(df) for df in samp_list.values()]
    return pd.concat(mult, axis=1, keys=samp_list.keys())
```

The dependence of the correlation time on concentration--the change from ~11.2 ns to ~9.7 ns between 0.75 and 0.16 mM--implies that there may be transient oligomerization, ~15% population at most. This is consistent with [Alexandrescu et al. 1989](http://pubs.acs.org/doi/abs/10.1021/bi00427a028), though it is worth noting two things:

1. Their concentrations of SNase were truly high--up to ~8 mM, compared to which even our 'high' concentration (0.75 mM) is very low.
2. We don't see evidence of oligomers in SDS-PAGE experiments with our ∆+PHS SNase.

Under our low-concentration condition (0.16 mM protein), there is only a slight decrease in rotational correlation time from 9.7 to 8.9 ns. Given the inaccuracies inherent in the TRACT approach, I would not consider this change to be significant.

Regardless of the particulars of putative oligomerization, the above results from two concentrations at two pH conditions indicate that we should make no claim or physical interpretation of the apparent pH dependence of rotational correlation time.
